```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

## CRIAR GRAFO E ADICIONAR NÓS 

```python
# Criar o grafo 
G = nx.Graph()

# Adicionar edges and nodes 
G.add_edge(1,2)
G.add_edge(2,3, weight=0.9)
G.add_edge("A", "B")
G.add_edge("B", "B")
G.add_node("C")
G.add_node(print)
G.add_node("Lícia")

# Mostrar o grafo 
nx.draw_spring(G, with_labels=True)
plt.show()
```

## CRIAR NÓS A PARTIR DE UMA LISTA


```python
edge_list = [(1,2), (2,3), (3,4), (3,5), (4,6), (6,7)]
G = nx.from_edgelist(edge_list)

# # Ou outra forma
G = nx.Graph()
G.add_edges_from(edge_list)

# Mostrar o grafo 
nx.draw_spring(G, with_labels=True)
plt.show()
```

## CRIAR GRAFOS A PARTIR DE UMA LISTA DE ADJACÊNCIA

```python

# # Podemos mostrar a matriz de adjacencia de um grafo
# print(nx.adjacency_matrix(G))

# Criar matriz de adjacência
matriz = np.array([
    [0,1,0],
    [1,1,1],
    [0,0,0]
])
G = nx.from_numpy_array(matriz)

# Mostrar o grafo 
nx.draw_spring(G, with_labels=True)
plt.show()
```
